```python
#sell for reading the file and save it in dictionary
def main():
    
    try:
        fileName = input("write the file name which you want to read : ")
        inFile = open(fileName,"r")
        lines = inFile.readlines()      #use it to find the number of lines

        info = []         #list for every line 
        course1 = []         #list for the courses of one student
        student = []         # the final list for one student

        dataBase = {}         #the final dictionary as dataBase for the file 

        for oneLine in lines :  
            student = []         #temporary list to clear with each line
            courses = []         #same 
            info = oneLine.rstrip().split(";")       # rstrip to get rid of the \n
            course1 = info[0].rstrip().split(",")
            student.append(course1[0])       # get the name and ID and major in other List 
            student.append(course1[1])
            student.append(course1[2])
            course1.pop(0)              #removing the name and ID and major after transfer it
            course1.pop(0)
            course1.pop(0)
            courses.append(course1)                #the first course is not included in the below loop


            for oneCourse in range (1,len(info)):                 # loop to separate the courses into list

                course1 = info[oneCourse].rstrip().split(",")
                courses.append(course1)

            student.append(courses)
            dataBase[str(student[1])] = student                #put the keys of the dictionary as students ID
        inFile.close()
    

        remove = []
        for key in dataBase:                #finding the wrong IDs
            if len(key) != 9 or key[0] != "2" or key[8] != "0" :               #the rules for ID
                remove.append(key)

        for keys in remove :               #removing the wrong IDs
            dataBase.pop(keys)
        
        
    

#--------------------------------------------------------------------------------------------------------------
        
        choice = str("")               #the main menu 
        
        while choice != "9" :
            print("\n1. Display registered courses for a student and his GPA")
            print("2. Display students in a course ")
            print("3. Display students with GPA more or equal to an entered GPA")
            print("4. Add a course to a student")
            print("5. Drop a course from a student")
            print("6. Add a student info")
            print("7. Delete a student info ")
            print("8. Save data to the file ")
            print("9. quite")
            choice = input("choose number : ")
            
            if choice == "1" :
                studentInfo = studentsInformation()               
                studentList = studentInfo.coursesGpa(dataBase)
                if studentList == 0:                        # if the ID isn't in the dictionary
                    print("ID not found")
                else :
                    print("the student GPA is : %0.2f"%(studentList[0]))
                    print("the courses :")
                    for i in range (len(studentList[1])):
                        print(studentList[1][i][0])
                Exit = str(input("press Enter to go back to main menu"))
                while Exit != "":
                    Exit = str(input("press Enter to go back to main menu"))    
                        
                        
            elif choice == "2" :
                studentInfo = studentsInformation()
                s = studentInfo.studentInC(dataBase)
                Exit = str(input("press Enter to go back to main menu"))   # a loop to make the user press enter to quit option
                while Exit != "":
                    Exit = str(input("press Enter to go back to main menu"))

                    
                    
            elif choice == "3" :                  #option 3 to find student with GPA higher then the entered GPA
                wantedGpa = float(input("write the GPA : "))
                if wantedGpa > 4 or wantedGpa < 0 :                 #restrictions for the GPA
                    print("invalid input")
                else :
                    studentInfo = studentsInformation()                    
                    passedGpa = studentInfo.aboveGpa(dataBase,wantedGpa)
                    for value in range(0,int(len(passedGpa)),3):        #printing the returned list
                        print("name: %-20s ID: %-15s GPA: %3.2f "%(passedGpa[value],passedGpa[value+1],passedGpa[value+2]))
                Exit = str(input("press Enter to go back to main menu"))
                while Exit != "":
                    Exit = str(input("press Enter to go back to main menu"))
                        
                        
            elif choice == "4" :
                studentInfo = studentsInformation()               
                dataBase = studentInfo.addCourse(dataBase)    #calling the class studentInfo
                Exit = str(input("press Enter to go back to main menu"))
                while Exit != "":
                    Exit = str(input("press Enter to go back to main menu"))



            elif choice == "5" :
                studentInfo = studentsInformation()               
                dataBase = studentInfo.dropCourse(dataBase)
                Exit = str(input("press Enter to go back to main menu"))
                while Exit != "":
                    Exit = str(input("press Enter to go back to main menu"))
                
                
                
            
            elif choice == "6" :
                studentInfo = studentsInformation()
                s = studentInfo.addInfo(dataBase)
                Exit = str(input("press Enter to go back to main menu"))
                while Exit != "":
                    Exit = str(input("press Enter to go back to main menu"))
                
                
                
            elif choice == "7" :
                studentInfo = studentsInformation()
                s = studentInfo.deleteInfo(dataBase)
                Exit = str(input("press Enter to go back to main menu"))
                while Exit != "":
                    Exit = str(input("press Enter to go back to main menu"))
            
            
            
            elif choice == "8" :
                studentInfo = studentsInformation()
                s = studentInfo.SaveInfo(fileName,dataBase)
                print("the changes saved successfully.")
                Exit = str(input("press Enter to go back to main menu"))
                while Exit != "":
                    Exit = str(input("press Enter to go back to main menu"))
                
                
        
    except FileNotFoundError as exception:            #exception if the file is not found
        print("Error:",str(exception))
    print("-------DONE-------")
    
#=================================================================================================
class studentsInformation :
    
    def __init__(self) :
        self._coursesGpa = 0  #the initials for the functions to return 
        self._aboveGpa = 0.00
        self._addCourse = 0
        self._dropCourse = 0
        self._coursenum = 0   # a vriable to check opotion 2
        self._file = ""
   
 #for option 1 ==================================================================
    
    def coursesGpa(self,list) :
        
        id = str(input("write the student ID : "))
        if id in list :                             
            if list[id][3] != [[]] :                      # if statement to prevent divition by zero
                
                self._totalCredits = 0
                self._totalPoints = 0.0
                for numOfCourses in range(len(list[id][3])) :                 #the loop to calculate the GPA
                    self._totalCredits = self._totalCredits + int(list[id][3][numOfCourses][2])
                    if list[id][3][numOfCourses][3].upper() == "A+" :
                        self._totalPoints += 4*int(list[id][3][numOfCourses][2])           #the grades weight
                    elif list[id][3][numOfCourses][3].upper() == "A" :
                        self._totalPoints += 3.75*int(list[id][3][numOfCourses][2])
                    elif list[id][3][numOfCourses][3].upper() == "B+" :
                        self._totalPoints += 3.5*int(list[id][3][numOfCourses][2])
                    elif list[id][3][numOfCourses][3].upper() == "B" :
                        self._totalPoints += 3*int(list[id][3][numOfCourses][2])
                    elif list[id][3][numOfCourses][3].upper() == "C+" :
                        self._totalPoints += 2.5*int(list[id][3][numOfCourses][2])
                    elif list[id][3][numOfCourses][3].upper() == "C" :
                        self._totalPoints += 2*int(list[id][3][numOfCourses][2])
                    elif list[id][3][numOfCourses][3].upper() == "D+" :
                        self._totalPoints += 1.5*int(list[id][3][numOfCourses][2])
                    elif list[id][3][numOfCourses][3].upper() == "D" :
                        self._totalPoints += 1*int(list[id][3][numOfCourses][2])
                    elif list[id][3][numOfCourses][3].upper() == "F" :
                        self._totalPoints += 0*int(list[id][3][numOfCourses][2])
                self._coursesGpa = self._totalPoints/self._totalCredits
                return self._coursesGpa, list[id][3]                        
            else :
                self._coursesGpa = 0.00
                return self._coursesGpa,[["the student has no courses"]]
        else :
            return self._coursesGpa   #if its not in the dictionary
                
            
            
 #for option 2 =====================================================================

    def studentInC(self,list) :
        course = str(input("write the course code : "))
        course = course.upper()
        print("students in course :")
        for id in list :           # for loop to check for every student
            x = list[id][3]
            for i in x :
                if course in i:    # for loop to check for every course per student
                    self._coursenum += 1   
                    print("name: %-20s ID: %-15s"%(list[id][0],list[id][1]))
                    
        if self._coursenum == 0 : # to check if no student has the course
            print("ERROR : no student has this course")
            
            
            
        
 #for option 3  ====================================================================         
            
    def aboveGpa(self,list,wantedGpa):
        
        self._passedGpa = []
        for studentID in list :                         #the loop to calculate the GPA for all students
            if list[studentID][3] != [[]] :                      # if statement to prevent divition by zero
                self._totalCredits = 0
                self._totalPoints = 0.0
                for numOfCourses in range(len(list[studentID][3])) :
                    self._totalCredits = self._totalCredits + int(list[studentID][3][numOfCourses][2])   #the same as option 1
                    if list[studentID][3][numOfCourses][3].upper() == "A+" :
                        self._totalPoints += 4*int(list[studentID][3][numOfCourses][2])
                    elif list[studentID][3][numOfCourses][3].upper() == "A" :
                        self._totalPoints += 3.75*int(list[studentID][3][numOfCourses][2])
                    elif list[studentID][3][numOfCourses][3].upper() == "B+" :
                        self._totalPoints += 3.5*int(list[studentID][3][numOfCourses][2])
                    elif list[studentID][3][numOfCourses][3].upper() == "B" :
                        self._totalPoints += 3*int(list[studentID][3][numOfCourses][2])
                    elif list[studentID][3][numOfCourses][3].upper() == "C+" :
                        self._totalPoints += 2.5*int(list[studentID][3][numOfCourses][2])
                    elif list[studentID][3][numOfCourses][3].upper() == "C" :
                        self._totalPoints += 2*int(list[studentID][3][numOfCourses][2])
                    elif list[studentID][3][numOfCourses][3].upper() == "D+" :
                        self._totalPoints += 1.5*int(list[studentID][3][numOfCourses][2])
                    elif list[studentID][3][numOfCourses][3].upper() == "D" :
                        self._totalPoints += 1*int(list[studentID][3][numOfCourses][2])
                    elif list[studentID][3][numOfCourses][3].upper() == "F" :
                        self._totalPoints += 0*int(list[studentID][3][numOfCourses][2])
                self._coursesGpa = self._totalPoints/self._totalCredits
                
            else :
                self._coursesGpa = 0.00            #if the courses list is empty
                    

            if self._coursesGpa >= wantedGpa :                   #sorting results in list
                self._passedGpa.append(list[studentID][0])
                self._passedGpa.append(list[studentID][1])
                self._passedGpa.append(self._coursesGpa)
                
                

                
        return self._passedGpa   
    
 #for option 4 ========================================================================

    def addCourse(self,list) :
        id = str(input("write the student's id : "))  #asking for the student Id
        if id in list :                                             #see if the ID in the dictionary
            courseCode = str(input("write the course code : "))
            courseName = input("write the course name : ")
            credits = int(input("write the course credit hours : "))
            if credits > 4 or credits < 1 :
                print("wrong input : credit hours should be more than 1 and less than or equal 4")
                credits = int(input("write the course credit hours : "))
            grade = str(input("write the grade in the course : "))
            while grade[0].upper() not in "ABCDF":
                print("wrong input : grade should be in these alphabet A,B,C,D,F")
                grade = input("write the grade in the course : ")
            
            courseInfo = []             # empty list for the course info
            courseInfo.append(courseCode.upper())
            courseInfo.append(courseName)
            courseInfo.append(credits)
            courseInfo.append(grade.upper())
            list[id][3].append(courseInfo)       #add the course as list
            print("course added successfully")
                
        else :
            print("ID is not found")
        
        return list                    #returning the new list 

 #for option 5 ========================================================================


    def dropCourse(self,list) :
        id = str(input("write the student ID : "))
            
        if id in list :                               #checking the ID first
            if list[id][3] == [[]] :                              #checking the courses if it is empty
                print("there is no courses to drop")
                return list
            else :
                courseCode = str(input("write the course code : ")).upper()       #course code in upper characters
                y = 0
                for courses in range(len(list[id][3])) :             #checking if the course in the list 
                    if courseCode == list[id][3][courses][0] :
                        z = courses           # a mark to know the number of the course in the list if found 
                    else :
                        y += 1                   # a mark to know if the course not found
                if y == len(list[id][3]):           #checking the marks
                    print("course not found")
                else :
                    list[id][3].pop(z)
                    print("the course has been dropped successfully")
                return list                       
        else :
            print("ID not found")
            return list                    #returning the list to go back to the main function
        
 #for option 6 ========================================================================

    def addInfo (self,list) :
        id = str(input("write the student ID : ")) 
        if len(id) != 9 or id[0] != "2" or id[len(id)-1]!= "0": #if statment to check id validity
            print("ERROR : id is not valid")
        elif id in list:
            print("ERROR : id already exist")   #if statment to check if id already exist
        else:
            Name = str(input("write the student name : "))  # getting student info
            Major = str(input("write the student Major: "))
            list[id] = [Name,id,Major,[]]  #appending student info into our list
            print("student Info has been added successfully")   #message shows function succes
            
            
 #for option 7 ========================================================================


    def deleteInfo(self,list) :
        id = str(input("write the student ID : "))
        if id in list :  # if statment to check id in list
            list.pop(id)
            print("student Info has been deleted successfully")
        else: 
            print("ERROR : id doesn't exist") #message if there is no student with this id 
            

            
 #for option 8 ========================================================================
    
    
    def SaveInfo(self,file,list) :
        self._file = open(file,"w")  # opening the same read file for writing
        for id in list:         # for loop to write every student info
            self._file.write(list[id][0]+","+list[id][1]+","+list[id][2])
            if list[id][3] != [[]]:  # to check if the courses not empty
                self._file.write(",")  # writing comma before courses if it is not embty
                for course in range(len(list[id][3])):  # for loop to get every course per student
                    for infos in range(len(list[id][3][course])): # for loop to get info for every course
                        self._file.write(str(list[id][3][course][infos]))
                        if infos != len(list[id][3][course])-1:   # to check if info for course is the last to avoid writing comma
                            self._file.write(",")
                    if course != len(list[id][3])-1:   # to check if course info finished to write semi-column
                        self._file.write(";")
                     
                    
            self._file.write("\n")   # a new line
        self._file.close()  #closing file to save changes
        

main()
```
